___

```python
# объединяем таблицы с информацией по курсам и с оценками студентов

students_assessment_extended = students_assessment.merge(assessments, on='id_assessment')
```

```python
# просмотрим таблицу

students_assessment_extended.head()
```

```python
'''
в объединенной таблице фильтруем:
1. Отображение типа теста только экзамены, сданные с оценкой выше 40
2. Группируем по студентам
3. Подсчитываем сколько каждый студент сдал экзаменов
'''

df = students_assessment_extended\
    .query("assessment_type == 'Exam' and score >= 40")\
    .groupby('id_student', as_index=False)\
    .agg({'code_module':'count'})\
    .rename(columns={'code_module':'passed_courses_count'})\
    .sort_values('passed_courses_count')
```

## Задание 2.1<a name="task_2_1"></a>
Найти самый сложный и самый легкий курс
___

**находим количество успешных попыток сдать экзамен**

Для этого:
1. фильтруем по типу теста 'Exam' и с оценкой более 40
2. группируем по курсу
3. подсчитываем количество оценок


```python
succesful_exams_count_by_course = students_assessment_extended\
                                    .query("assessment_type == 'Exam' and score >= 40")\
                                    .groupby('code_module', as_index=False)\
                                    .agg({'score':'count'})\
                                    .rename(columns={'score':'successful'})
```

```python
succesful_exams_count_by_course
```

**находим общее количество попыток сдать экзамен**

Для этого:
1. фильтруем по типу теста 'Exam'
2. группируем по курсу
3. подсчитываем количество оценок


```python
succesful_exams_count_by_course['total_tries'] = students_assessment_extended\
        .query("assessment_type == 'Exam'")\
        .groupby('code_module', as_index=False)\
        .agg({'score':'count'})\
        .score
```

```python
succesful_exams_count_by_course
```

**рассчитываем завершаемость экзамена**

```python
succesful_exams_count_by_course['success_percentage'] = \
            succesful_exams_count_by_course.successful / succesful_exams_count_by_course.total_tries * 100
```

```python
succesful_exams_count_by_course.sort_values('success_percentage')
```

```python
# присваиваем мин и макс для экзаменов с самой низкой и самой высокой завершаемостью

max = succesful_exams_count_by_course.success_percentage.max()
min = succesful_exams_count_by_course.success_percentage.min()
```

```python
# присваиваем переменный, которые показывают названия для курсов с самой низкой и самой высокой завершаемостью

most_easy_course = succesful_exams_count_by_course.loc[succesful_exams_count_by_course['success_percentage']  == max].code_module.to_list()[0]
most_difficult_course = succesful_exams_count_by_course.loc[succesful_exams_count_by_course['success_percentage']  == min].code_module.to_list()[0]
```

### Ответ на задание 2.1<a name="task_2_1_answer"></a>

```python
print(f'Самую низкую завершаемость {round(min, 2)} имеет курс {most_difficult_course}')
print(f'Самую высокую завершаемость {round(max, 2)} имеет курс {most_easy_course}')
```

## Задание 2.2<a name="task_2_2"></a>
Найти самый сложный и самый легкий экзамены в рамках курса
___

**находим количество успешных попыток сдать экзамен**

Для этого:
1. фильтруем по типу теста 'Exam' и с оценкой более 40
2. группируем по курсу и по номеру теста
3. подсчитываем количество оценок


```python
succesful_exams_count_by_course_and_assessment = students_assessment_extended\
                        .query("assessment_type == 'Exam' and score >= 40")\
                        .groupby(['code_module','id_assessment'], as_index=False)\
                        .agg({'score':'count'})\
                        .rename(columns={'score':'successful'})
succesful_exams_count_by_course_and_assessment
```

**находим общее количество попыток сдать экзамен**

Для этого:
1. фильтруем по типу теста 'Exam'
2. группируем по курсу и по номеру теста
3. подсчитываем количество оценок

```python
succesful_exams_count_by_course_and_assessment['total_tries'] = students_assessment_extended\
        .query("assessment_type == 'Exam'")\
        .groupby(['code_module', 'id_assessment'], as_index=False)\
        .agg({'score':'count'})\
        .score
succesful_exams_count_by_course_and_assessment
```

**рассчитываем завершаемость экзамена**

```python
succesful_exams_count_by_course_and_assessment['success_percentage'] = \
            succesful_exams_count_by_course_and_assessment.successful / \
            succesful_exams_count_by_course_and_assessment.total_tries * 100
```

```python
succesful_exams_count_by_course_and_assessment.sort_values('success_percentage')
```

**формируем таблицу, для разбивки по курсам**

```python
succesful_exams_count_by_course_and_assessment = succesful_exams_count_by_course_and_assessment.pivot(index = 'id_assessment',
                                                     columns = 'code_module',
                                                     values = 'success_percentage')
```

```python
succesful_exams_count_by_course_and_assessment
```

### Ответ на задание 2.2<a name="task_2_2_answer"></a>

```python
'''
для вывода ответа использую проход по столбцам и определение минимума, максимума завершаемости
также определяется id экзамена в рамках курса, которые имеет мин/макс завершаемость
это сделано, чтобы данный скрипт работал не только для вот этого конкретного набора данных
если набор данных изменится, то скрипт будет считать и выводить ответ мин/макс для любого другого количества курсов и экзаменов
'''

for column in succesful_exams_count_by_course_and_assessment.columns:
    min = succesful_exams_count_by_course_and_assessment[column].min()
    min_id_assessment = succesful_exams_count_by_course_and_assessment.loc[succesful_exams_count_by_course_and_assessment[column] == min][column].index[0]
    max = succesful_exams_count_by_course_and_assessment[column].max()
    max_id_assessment = succesful_exams_count_by_course_and_assessment.loc[succesful_exams_count_by_course_and_assessment[column] == max][column].index[0]
    print(f'в курсе {column} самый сложный экзамен с id = {min_id_assessment}. Его завершаемость {round(min, 2)}')
    print(f'в курсе {column} самый легкий экзамен с id = {max_id_assessment}. Его завершаемость {round(max, 2)}')
    print('')
    
```

функция *cohort_analysis* принимает три аргумента:

2. Датасет с информацией о результатах тестов студентов. 
3. Список семестров, по которым проводится анализ

На выходе функция выдает таблицу, в которой посчитаны процент завершаемости курсов и средние сроки сдачи курсов

```python
def cohort_analysys(tests, students_tests, code_presentation):
# объединяем два датасета по общему полю id_assessment
    tests_and_students = tests.merge(students_tests, on='id_assessment')
# фильтруем по семестрам, по которым необходимо провести анализ
    tests_and_students_semester = tests_and_students[(tests_and_students.code_presentation.isin(code_presentation))]

#считаем общее количество попыток сдать курс
    total_exams = tests_and_students_semester\
        .query("assessment_type == 'Exam'")\
        .groupby('code_presentation', as_index=False)\
        .agg({'score':'count'})\
        .rename(columns={'score':'total_attempts'})

## считаем количество успешно сданных экзаменов по курсу
    successful_exams = tests_and_students_semester\
        .query("assessment_type == 'Exam' and score >= 40")\
        .groupby('code_presentation', as_index=False)\
        .agg({'score':'count'})\
        .rename(columns={'score':'successful_attempts'})

#объединим в одну таблицу и посчитаем успешный процент сдачи

    exams = total_exams.merge(successful_exams, on='code_presentation')
    exams['success_percentage'] = exams.successful_attempts / exams.total_attempts * 100

#считаем среднюю длительность сдачи экзаменов
    exams_date_submitted = tests_and_students_semester\
        .query("assessment_type == 'Exam' and score >= 40")\
        .groupby('code_presentation', as_index=False)\
        .agg({'date_submitted':'mean'})\
        .rename(columns={'date_submitted':'avg_date_submitted'})

#объединяем в итоговую таблицу и удаляем промежуточные колонки
    exams = exams.merge(exams_date_submitted, on='code_presentation')
    exams = exams.drop(columns=['total_attempts','successful_attempts'])
    exams.sort_values('success_percentage')

    return exams
```

### Итоговая таблица RFM <a name="task_6_rfm_summary_table"></a>

|                     | Дата сдачи теста | R | Количество сданных экзаменов | F | Среднее количество баллов за тест | M |
|---------------------|------------------|---|------------------------------|---|-----------------------------------|---|
| Граница 1 категории | 228 - 235        | 1 | 2                            | 1 | 81-100                            | 1 |
| Граница 2 категории | 236 - 243        | 2 | 1                            | 2 | 40-80                             | 2 |
| Граница 3 категории | 244 - 286        | 3 | 0                            | 3 | 0-39                              | 3 |

```python
# строим таблицу с данными необходимыми для сегментирования
exams = students_exams\
    .query("assessment_type == 'Exam'")\
    .groupby('id_student', as_index=False)\
    .agg({'date_submitted':'mean'
         , 'passed':'sum'
         , 'score':'mean'})
exams.head()
```

```python
# проводим категоризацию пользователей по RFM

exams = exams.assign(recency=pd.cut(exams.date_submitted, bins=[228, 235, 243, 286], labels=[1, 2, 3] ))
exams = exams.assign(frequency=pd.cut(exams.passed, bins=[0, 0.1, 1.1, 2], labels=[3, 2, 1], right=True, include_lowest=True))
exams = exams.assign(monetary=pd.cut(exams.score, bins=[0, 39, 80, 100], labels=[3, 2, 1], right=True, include_lowest=True))
```

```python
exams.head()
```

```python
# объединим три строки в одну, чтобы получить RFM категорию

exams['rfm'] = exams['recency'].astype(str) + exams['frequency'].astype(str) + exams['monetary'].astype(str)
```

```python
# для красоты можно удалить промежуточные колонки
rfm = exams.drop(columns=['date_submitted', 'passed', 'score', 'recency', 'frequency', 'monetary'])
```
